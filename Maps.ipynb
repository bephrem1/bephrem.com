```python
!pip install folium
import folium
import requests
import pandas
import random
```

```python
arrest_table = pandas.read_csv("http://www.hcbravo.org/IntroDataSci/misc/BPD_Arrests.csv")

arrest_table = arrest_table[pandas.notnull(arrest_table["Location 1"])]

arrest_table["lat"], arrest_table["long"] = arrest_table["Location 1"].str.split(",").str
arrest_table["lat"] = arrest_table["lat"].str.replace("(", "").astype(float)
arrest_table["long"] = arrest_table["long"].str.replace(")", "").astype(float)

arrest_table = arrest_table.head(1000)
arrest_table
```

```python
area_to_crime = {}

for index, row in arrest_table.iterrows():
    neighborhood = row['neighborhood']

    if neighborhood in area_to_crime:
        area_to_crime[neighborhood].append(row)
    else:
        area_to_crime[neighborhood] = []
        area_to_crime[neighborhood].append(row)
```

```python
def random_color():
    r = lambda: random.randint(0,255)
    return '#%02X%02X%02X' % (r(),r(),r())

map_osm = folium.Map(location=[39.29, -76.61], zoom_start=11)

for key, rows in area_to_crime.items():
    folium.CircleMarker(
        location=[rows[0]['lat'], rows[0]['long']],
        radius=len(rows)*1,
        popup="{} - Total Arrests {}".format(key, len(rows)),
        color=random_color(),
        fill=True,
        fill_color=random_color()
    ).add_to(map_osm)

map_osm
```

```python
print(
'''
The intent in the map display is to demonstrate relative total arrests for each neighborhood area in the dataset provided. The first 1000 entires were taken for processing purposes. We can see roughly visually that the top 4 neighborhoods with arrests are Sandtown-Winchester, Central Park Heights, Penn North, & McElderry Park. We can also loosely see that the deeper once goes into the city, the higher the frequency of arrest in each neighborhood.
'''
)
```
